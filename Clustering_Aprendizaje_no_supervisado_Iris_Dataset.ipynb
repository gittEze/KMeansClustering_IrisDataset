```python
import pandas as pd
df = pd.read_csv('IRIS.csv')

df.head()
```

```python
print("Cantidad de filas antes de eliminar duplicado", df.shape)

df.drop_duplicates(inplace=True)
print("Cantidad de filas despus de eliminar duplicado", df.shape)
```

```python
#Verificar si hay valores nulos
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
X_std = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_std)
df['cluster'] = kmeans.labels_
```

```python
#print(df[['species','cluster']].head(60))

print()

#print(df[df['species']=='Iris-virginica'])
print(df[df['species']=='Iris-virginica'][['species','cluster']])

print()

print(df[df['species']=='Iris-virginica'][['species','cluster']].sum())
```
